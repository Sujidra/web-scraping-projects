```python
import requests
import pandas as pd
import time
from lxml import html
from selenium import webdriver
import lxml, lxml.html
driver = webdriver.Chrome("C:\\Users\sujidra\Documents\chromedriver_win32\chromedriver")
```

```python
title=[]
price=[]
processor=[]
RAM=[]
OS=[]
```

```python
for i in range(0,24):
    website='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_0_1_na_na_pr&otracker1=AS_QueryStore_OrganicAutoSuggest_0_1_na_na_pr&as-pos=0&as-type=HISTORY&suggestionId=laptop&requestId=73116be7-cca3-4757-bf39-150d59b3fbeb&page='+str(i)
    driver.get(website)
    tree = html.fromstring(driver.page_source)
    titles=tree.xpath("//div[@class='col col-7-12']/div[@class='_3wU53n']")
    prices=tree.xpath("//div[@class='_6BWGkk']/div[@class='_1uv9Cb']/div[@class='_1vC4OE _2rQ-NK']")
    processors=tree.xpath("//div[@class='col col-7-12']/div[@class='_3ULzGw']/ul[@class='vFw0gD']/li[@class='tVe95H'][1]")
    RAMs=tree.xpath("//ul[@class='vFw0gD']/li[@class='tVe95H'][2]")
    OSs=tree.xpath("//div[@class='_3ULzGw']/ul[@class='vFw0gD']/li[@class='tVe95H'][3]")
    for a in range(len(titles)-1):
        tit= html.tostring(titles[a])
        pri=html.tostring(prices[a])
        proc=html.tostring(processors[a])
        Ram=html.tostring(RAMs[a])
        os=html.tostring(OSs[a])
        tit="".join(map(chr, tit))
        pri="".join(map(chr,pri))
        proc="".join(map(chr,proc))
        Ram="".join(map(chr,Ram))
        os="".join(map(chr,os))
        tit=tit.split('>')
        pri=pri.split('>')
        Ram=Ram.split('>')
        proc=proc.split('>')
        os=os.split('>')
        tit=tit[1]
        pri=pri[1]
        proc=proc[1]
        Ram=Ram[1]
        os=os[1]
        tit=tit.split("<")
        pri=pri.split("<")
        proc=proc.split("<")
        Ram=Ram.split("<")
        os=os.split("<")
        tit=tit[0]
        pri=pri[0]
        proc=proc[0]
        Ram=Ram[0]
        os=os[0]
        pri=pri.split(";")
        pri=pri[1]
        title.append(tit)
        price.append(pri)
        processor.append(proc)
        RAM.append(Ram)
        OS.append(os)
    time.sleep(5)
    print(title)
    print(price)
    print(processor)
    print(RAM)
    print(OS)
```

```python
df = pd.DataFrame({'Product Name':title,'Price':price,"Processor":processor, "RAM":RAM ,"OS":os}) 
df.to_csv('laptops.csv', index=False, encoding='utf-8')
```
